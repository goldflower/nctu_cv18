# part1: sift for finding interest points with features

```python
def rgb2gray(rgb): # map rgb to gray for plt default color space
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
def get_img(f_name, rotate=False):
    """ load the image
    @args:
        f_name: file path
        rotate: true for rotate -90 (i.e., 90 clockwise)
    @returns:
        img(np array)
    """
    img = plt.imread(f_name) * 255 # sift need 0 - 255
    if rotate:
        # for jpg only
        img = transform.rotate(img, -90, resize=True, clip=False, preserve_range=True) * 255
    img = img.astype(np.uint8)        
    return img
```

```python
img1 = get_img('image1.png', rotate=False)
img2 = get_img('image2.png', rotate=False)
```

```python
sift = cv2.xfeatures2d.SIFT_create()

points1, features1 = sift.detectAndCompute(img1,None)
points2, features2 = sift.detectAndCompute(img2,None)
```

```python
def plot_sift_graph(img, points, f_name='img1_sift.png'):
    sift_point = [p.pt for p in points]
    circles = [Circle((p[0], p[1]), 5, facecolor='None', edgecolor='r', lw=0.1) for p in sift_point]
    fig, ax = plt.subplots()
    ax.imshow(img)
    for c in circles:
        ax.add_patch(c)
    ax.set_xticks([])  
    ax.set_yticks([])    
    plt.savefig(f_name, dpi=500, bbox_inches='tight', frameon=False)
    plt.show()
    plt.close()
```

```python
plot_sift_graph(img1, points1, 'img1_sift.png')
plot_sift_graph(img2, points2, 'img2_sift.png')
```

# part2: using ratio distance to find point pairs

```python
class MATCH():
    def __init__(self, src_index, dest_index, distance):
        self.src_index=src_index
        self.dest_index=dest_index
        self.distance=distance
```

```python
def naive_ratio_distance_mathcer(src_features, dest_features):
    """ratio distance
    norm2(f1-f2) / norm2(f1-f2')
    """
    matches = []
    feature_distances = np.zeros(dest_features.shape[0])
    for i, feature in zip(range(len(src_features)), src_features):
        feature_distances = np.sqrt(np.sum(np.square(np.subtract(src_features[i,:], dest_features)), axis=1))
        top2 = np.argsort(feature_distances)[:2]
        dist = feature_distances[top2[0]] / feature_distances[top2[1]]
        M = MATCH(i, top2[0], dist)
        matches.append(M)
    return matches
```

```python
def match(features1, features2, method='naive'):
    if method == 'naive':
        matches = naive_ratio_distance_mathcer(features1, features2)
    elif method == 'flann_knn':
        raise NotImplementedError
    matches = sorted(matches, key = lambda x: x.distance)
    return matches
```

```python
def draw_matched(img1, img2, points1, points2, good=None, f_name='matched.png'):
    
    if good is None:
        good = [m for m in matches if m.distance < 0.8]
    # for simplicity, we just resize the two image into the same size
    r_scale, c_scale = img2.shape[0] / img1.shape[0], img2.shape[1] / img1.shape[1]
    concate_img = misc.imresize(img1, (img2.shape[0], img1.shape[1], img1.shape[2]))
    concate_img = np.hstack([concate_img, img2])

    fig, ax = plt.subplots()
    plt.imshow(concate_img)
    for match in good:
        src_point = list(points1[match.src_index].pt)
        src_point[0] = r_scale * src_point[0]
        src_point[1] = c_scale * src_point[1]
        dest_point = points2[match.dest_index].pt
        # draw circles
        c1 = Circle((src_point[0], src_point[1]), 5, facecolor='None', edgecolor='r', lw=0.1)
        c2 = Circle((img1.shape[1] + dest_point[0], dest_point[1]), 5, facecolor='None', edgecolor='r', lw=0.1)
        ax.add_patch(c1)
        ax.add_patch(c2)
        ax.plot([src_point[0], img1.shape[1] + dest_point[0]], [src_point[1], dest_point[1]], lw=0.2)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.savefig(f_name, dpi=500, bbox_inches='tight', frameon=True)
    plt.show()
    plt.close()
```

# part3: using RANSAC to compute homography matrix

```python
def to_homogeneous(points):
    """ change points to it's homogeneous coordinate form
    args:
        points(list): [(pt1_x, pt1_y), (pt2_x, pt2_y), ...]
    returns:
        homo_points(np array): np.array([[pt1_x, pt2_x, ...], 
                                         [pt1_y, pt2_y, ...],
                                         [    1,     1, ...]])
    """
    points_arr = []
    for p in points:
        points_arr.extend([p[0], p[1]])
    points_arr = np.array(points_arr)
    points_arr = points_arr.reshape((-1, 2)).T # [[pt1_x, pt2_x, ...]
                                               #  [pt1_y, pt2_y, ...]]
    points_arr = np.vstack([points_arr, np.ones(points_arr.shape[1])])
    return points_arr
```

```python
good = [m for m in matches if m.distance < 0.8]
src_points = [points1[match.src_index].pt for match in good]
dest_points = [points2[match.dest_index].pt for match in good]
# homo = homogeneous coordinate, no more meanings
src_homo = to_homogeneous(src_points)
dest_homo = to_homogeneous(dest_points)
```

```python
def compute_homography(src_points, dest_points):
    """ map src_points to dest_points, hope H(src_points) = dest_points
    """
    A = np.zeros((2*src_points.shape[1], 9))
    for i in range(src_points.shape[1]):
        x, y = src_points[0, i], src_points[1, i]
        x_p, y_p = dest_points[0, i], dest_points[1, i]
        A[2*i, :] = np.array([-x,-y,-1,0,0,0,x*x_p,y*x_p,x_p])
        A[2*i+1,:] = np.array([0,0,0,-x,-y,-1,x*y_p,y*y_p,y_p])

    U,S,V = np.linalg.svd(A)
    H = V[-1].reshape((3,3))

    return H / H[-1,-1] # we know the right bottom of homography matrix is 1
```

```python
def compute_distance(src_homo, dest_homo, H):
    """ Euclien distance between dest_homo and transformed homo from src_homo by H
    """
    transformed = np.dot(H, src_homo)
    transformed /= transformed[-1, :]
    distances = np.sqrt(np.sum(np.square(dest_homo - transformed), 0))
    return distances
```

```python
def find_H_via_ransac(src_homo, dest_homo, nb_sample=4, N=5000, distance_threshold=100, adjust_N=False):
    """ envolke ransac algorithm
    @Args:
        src_homo(np array): 3 x n homogeneous 
        dest_homo(np array): 3 x n homogeneous         
        nb_sample(int): for computing H
        N(float): Choose N so that, with probability p, at least one random sample is
                  free from outliers (e.g. p=0.99) (outlier ratio: e )
        distance_threshold: distinguish outlier/inlier
    @returns:
        H(np array): 3x3 homography matrix
    """
    iterations = 0
    best_H = None
    best_distance = np.inf
    best_inlier_index = []
    p = 0.99 # for adjust N
    
    while iterations < N:
        random_order = np.arange(src_homo.shape[1])
        np.random.shuffle(random_order)
        
        # random sample index and the corresponded coordinates
        inliner_index = random_order[:nb_sample]
        else_index = random_order[nb_sample:]
        
        batch_src_homo = src_homo[:, inliner_index]
        batch_else_src_homo = src_homo[:, else_index]
        batch_dest_homo = dest_homo[:, inliner_index]
        batch_else_dest_homo = dest_homo[:, else_index]  
        
        # compute H, check distance, if it's better, update
        temp_H = compute_homography(batch_src_homo, batch_dest_homo)
        distances = compute_distance(batch_else_src_homo, batch_else_dest_homo, temp_H)
        other_inliner_index = else_index[distances < distance_threshold]
        
        all_inliner_index = np.hstack([inliner_index, other_inliner_index])
        all_inliner_src_homo = src_homo[:, all_inliner_index]
        all_inliner_dest_homo = dest_homo[:, all_inliner_index]
        this_H = compute_homography(all_inliner_src_homo, all_inliner_dest_homo)
        mean_distance = np.mean(compute_distance(all_inliner_src_homo, all_inliner_dest_homo, this_H))
        if len(all_inliner_index) > len(best_inlier_index):
            # tried  nb_inliners, smaller mean_distance, the results are similar
            best_H = this_H
            best_distance = mean_distance
            best_inlier_index = all_inliner_index     
            print('better!!', len(best_inlier_index))
        iterations+=1
        if adjust_N:
            e = 1.0 - len(all_inliner_index) / src_homo.shape[1]
            N = np.log(1-p+1e-10) / np.log(1-(1-e)**nb_sample+1e-10) # divide by zero is annoying
    return best_H, best_inlier_index
    
```

```python
H, index = find_H_via_ransac(src_homo, dest_homo, N=50000, adjust_N=True)
```
